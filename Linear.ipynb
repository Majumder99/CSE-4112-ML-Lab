```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m=1.2
c=2
#line points
X = np.array([i for i in range(-1, 50)])
Y_old = m*X + c
Y= np.random.normal(0,3,X.shape[0]) + Y_old
```

```python
m=1.2 
c=2 
#line points 
X = np.array([i for i in range(-1, 50)]) 
Y_old = m*X + c 
Y= np.random.normal(0,3,X.shape[0]) + Y_old
```

```python
plt.plot(X,Y_old,label="actual")
plt.scatter(X,Y,label="dataset")
plt.legend()
plt.show()
```

```python
def model(X,w,b):
    return X*w + b
```

```python
model(X,1,2)
```

```python
def loss(w,b):
    mse = (1/Y.shape[0])*np.sum((Y-model(X,w,b))**2)
    return mse
```

```python
loss(2,1)
```

```python
def grad_desc(lr = 1,epoch=10):
    w=0
    b=0
    for i in range(1,epoch):
        if i % 5 == 0:
            lr = lr/2
        #calculate derivative against parameters
        dw=(-2/Y.shape[0])*np.sum(X*(Y-model(X,w,b)))
        db=(-2/Y.shape[0])*np.sum((Y-model(X,w,b)))
        #dw=(1/Y.shape[0])*np.sum(-2*X*(Y-model(X,w,b)))
        
        #update parameters
        w = w - lr*dw
        b = b - lr*db
        
        print(i+1," Loss: ", loss(w,b),", w: ", w," ,b: ", b)
        
        #plot the lines
        Y_hat = w * X + b
        plt.plot(X,Y_hat, label="Epoch: "+str(i))
    plt.scatter(X,Y,label="dataset")
    plt.legend()
    plt.show()
    
    return w, b
        
```

```python
w,b= grad_desc(.005,20)
```
